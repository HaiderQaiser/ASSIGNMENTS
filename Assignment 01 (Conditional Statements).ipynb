1) A company decided to give bonus of 5% to employee if his/her year of service is more than 5 years.
Ask user for their salary and year of service and print the net bonus amount.

```python
salary = float(input("Please Enter your Salary:"))
experience = int(input("Please Enter your years of service: "))
percent_value = 0

if experience > 5:
    percent_value = (5 * salary)/100
    print(f"Your net bonus is: { percent_value + salary }")
else:
    print("Sorry, you can not get bonus!")
```

```python
num = int(input("Enter a number to check whether it is even or odd: "))

if (num % 2) == 0:
    print("The entered number is even!")
else:
    print("The entered number is odd!")
```

8. A school has following rules for grading system:

a. Below 25 - F

b. 25 to 45 - E

c. 45 to 50 - D

d. 50 to 60 - C

e. 60 to 80 - B

f. Above 80 - A

Ask user to enter marks and print the corresponding grade.

```python
marks = int(input("Enter your marks: "))

if marks >= 80:
    print("Your grade is A")
elif marks >= 60:
    print("Your grade is B")
elif marks >= 50:
    print("Your grade is C")
elif marks >= 45:
    print("Your grade is D")
elif marks >= 25:
    print("Your grade is E")
elif marks < 25:
    print("You are failed!")
```

9. Ask user to enter age, gender ( M or F ), marital status ( Y or N ) and then using following rules print their place of service.

if employee is female, then she will work only in urban areas.

if employee is a male and age is in between 20 to 40 then he may work in anywhere

if employee is male and age is in between 40 t0 60 then he will work in urban areas only.

And any other input of age should print "ERROR"

```python
age = int(input("Enter your age: "))
gender = input("Enter your Gender, M or F: ")
#marital_status = input("Enter your marital status, Yes or No:")

if gender.lower() == "f":
    print("You can only work in Urban Areas.")
elif gender.lower() == "m":
    if age >= 20 and age <= 40:
        print("You can work anywhere.")
    elif age >= 40 and age <= 60:
        print("You can only work in Urban Areas")
    else:
        print("Error")    
else:
    print("You typed something wrong!")
```
